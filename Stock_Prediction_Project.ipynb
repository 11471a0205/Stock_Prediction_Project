```python
apprcep_duptes = apprcepdta[apprcepdta.duplicated(keep=False)]

print("Duplicate rows in the DataFrame:", len(apprcep_duptes))
```

**Convert 'Date' column to datetime format**

```python
apprcepdta['Date'] = apprcepd_pdsn.to_datetime(apprcepdta['Date'])
```

```python
apprcepdta
```

```python

apprcepdta['Close/Last'] = apprcepdta['Close/Last'].str.replace('$', '', regex=False).astype(float)
apprcepdta['Open'] = apprcepdta['Open'].str.replace('$', '', regex=False).astype(float)
apprcepdta['High'] = apprcepdta['High'].str.replace('$', '', regex=False).astype(float)
apprcepdta['Low'] = apprcepdta['Low'].str.replace('$', '', regex=False).astype(float)
```

**Plotting the Closing Price:**

```python
import matplotlib.pyplot as apprcep_mplit
import seaborn as apprcep_slit
```

```python
apprcep_mplit.style.use('seaborn-v0_8-dark-palette')
apprcep_mplit.figure(figsize=(14, 7))

apprcep_slit.lineplot(y=apprcepdta["Close/Last"][:500], x=apprcepdta["Date"][:500], color='black')
apprcep_mplit.title('Apple stocks Over Time')
apprcep_mplit.xlabel('Date')
apprcep_mplit.ylabel('Closing Price')
apprcep_mplit.legend()
apprcep_mplit.show()
```
